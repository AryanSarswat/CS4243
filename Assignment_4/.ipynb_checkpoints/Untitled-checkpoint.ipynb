```python
def neighborhood_points(features, centroid, bandwidth):
    diff = features - centroid
    norm = np.linalg.norm(diff, axis=1)
    in_bandwidth = features[np.where(norm < bandwidth)]
    return in_bandwidth
```

```python
def mean_shift(features, bandwidth, gamma_h):
    prev_centroids = np.copy(features)
    temp_bandwidth = bandwidth
    while True:
        new_centroids = np.copy(prev_centroids)
        for i, x in enumerate(prev_centroids):
            neighbours = neighborhood_points(prev_centroids, x, temp_bandwidth)
            new_centroid = np.mean(neighbours)
            new_centroids[i] = new_centroid
        diff = np.sum(np.abs(new_centroids - prev_centroids))
        temp_bandwidth*= gamma_h
        if diff < 1:
            prev_centroids = new_centroids
            break
        else:
            prev_centroids = new_centroids
    
    prev_centroids = np.around(prev_centroids)
    cluster_centers = np.unique(prev_centroids, axis = 0)
    labels = np.zeros(features.shape[0]).astype(int)
    for i in range(len(prev_centroids)):
        l = np.argwhere((cluster_centers == prev_centroids[i]).all(1))
        labels[i] = int(l[0])
    clustering = {'cluster_centers_': cluster_centers, 'labels_': labels, 'bandwidth': bandwidth}
    return clustering
            
```
